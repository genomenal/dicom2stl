## Merge and Extract Mesh:

```python
objects_toextract = [x for x in splitted_mesh if not (x.faces.shape[0] == 66682 or x.faces.shape[0] == 34070)]
```

```python
len(objects_toextract)
```

```python
idx_of_objects_toextract = [0, 1, 2]
objects = []

for i, obj in enumerate(splitted_mesh):
    if i in idx_of_objects_toextract:
        objects.append(obj)
```

```python
objects
```

```python
vertice_list = [mesh.vertices for mesh in objects_toextract]
faces_list = [mesh.faces for mesh in objects_toextract]
faces_offset = np.cumsum([v.shape[0] for v in vertice_list])
faces_offset = np.insert(faces_offset, 0, 0)[:-1]

vertices = np.vstack(vertice_list)
faces = np.vstack([face + offset for face, offset in zip(faces_list, faces_offset)])

merged__meshes = trimesh.Trimesh(vertices, faces)
```
